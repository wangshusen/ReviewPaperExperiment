```python
import numpy
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
%matplotlib inline  

import sys
sys.path.append('../../main/kernel/')
import kernelFun 
import rbfKernelApprox


n_samples = 200
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.03)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
no_structure = numpy.random.rand(n_samples, 2), None
noisy_s_curve = datasets.make_s_curve(n_samples=n_samples, noise=0.05)
```

```python
X = noisy_circles[0]
y = noisy_circles[1]
colors = numpy.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = numpy.hstack([colors] * 20)

idx1 = (y==1)
idx0 = (y==0)
idx = numpy.array(range(200))
X1 = X[idx1, :]
X0 = X[idx0, :]
y1 = y[idx1]
y0 = y[idx0]
X = numpy.concatenate((X1, X0))
y = numpy.concatenate((y1, y0))
y[199] = 2


```

```python
matK = kernelFun.rbf(X, X, 0.3)
matU, vecS, _ = numpy.linalg.svd(matK, full_matrices=False)
k = 2
matXkpca = matU[:, 0:k] * numpy.sqrt(vecS[0:k]).reshape(1, k)

plt.scatter(matXkpca[:, 0], matXkpca[:, 1], color=colors[y].tolist(), s=10)
plt.show()

degree = sum(matK).reshape(200, 1)
```

```python
matKnew = numpy.zeros((200, 200))
matKnew[:, :] = matK
matKnew[199, 0:100] = matKnew[199, 0:100] - 0.07
matKnew[0:100, 199] = matKnew[0:100, 199] - 0.07
matKnew = (matKnew + matKnew) / 2
print(sum(matKnew[199, 0:100]))
matU, vecS, _ = numpy.linalg.svd(matKnew, full_matrices=False)
k = 2
matXkpca = matU[:, 0:k] * numpy.sqrt(vecS[0:k]).reshape(1, k)

plt.scatter(matXkpca[:, 0], matXkpca[:, 1], color=colors[y].tolist(), s=10)
plt.show()
 
vecS, _ = numpy.linalg.eig(matKnew)
print(min(vecS))
```

```python
degree = numpy.abs(numpy.sum(matKnew, axis=1))
degree = degree.reshape(200, 1)
matA = matKnew / numpy.sqrt(degree)
matA = matA / numpy.sqrt(degree).T
matA = (matA + matA.T) / 2

vecS, matV = numpy.linalg.eig(matA)
index = numpy.argsort(vecS)
vecS = vecS[index]
matV = matV[:, index]
print(vecS)

matF = matV[:, 197:200]
normalizer = numpy.sum(matF ** 2, axis=1)
matF = matF #/ numpy.sqrt(normalizer.reshape(200, 1))
plt.scatter(matF[:, 2], matF[:, 1], color=colors[y].tolist(), s=10)
plt.show()
```

```python
matD = numpy.sum(matKnew, axis=1)
matA = matKnew / matD.reshape(200, 1)

vecS, matV = numpy.linalg.eig(matA)
index = numpy.argsort(vecS)
vecS = vecS[index]
matV = matV[:, index]
print(vecS)

matF = matV[:, 197:200]
normalizer = numpy.sum(matF ** 2, axis=1)
matF = matF #/ numpy.sqrt(normalizer.reshape(200, 1))
plt.scatter(matF[:, 1], matF[:, 0], color=colors[y].tolist(), s=10)
plt.show()
```

```python
vecD = sum(matKnew) 
matA = numpy.diag(vecD) - matKnew

vecS, matV = numpy.linalg.eig(matA)
index = numpy.argsort(vecS)
vecS = vecS[index]
matV = matV[:, index]
print(vecS)

matF = matV[:, 0: 4]
plt.scatter(matF[:, 1], matF[:, 2], color=colors[y].tolist(), s=10)
plt.show()
```
